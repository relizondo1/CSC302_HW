```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

**Q1** Please use Pandas to read olympic_medals.csv and use parallel_categories function from plotly.express to visualize proportions of medal type for each gender from since year 2000. Please see the example in the Python notebook we walked through in the class.

```python
filename = '/content/drive/MyDrive/DATA/olympic_medals.csv'
df = pd.read_csv(filename)
df.head()
df_filtered = df[df['Year'] >= 2000]
df_filtered.head()
```

```python
fig = px.parallel_categories(df_filtered, dimensions=['Medal', 'Gender'],
                             color="Year",
                             color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```

Then, utilize stacked bar plot, to stack vote percentages for Trump, Clinton, Johnson, and Others. Please see 'pct_clinton',	'pct_trump',	'pct_johnson',	'pct_other' columns. Make sure that your x tick labels are those four states above.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# create DataFrame
df = pd.DataFrame({'High Temp': [28, 30, 34, 38, 45, 42,
                                 38, 35, 32, 28, 25, 21],
                   'Low Temp': [22, 26, 30, 32, 41, 38,
                                32, 31, 28, 22, 15, 15],
                   'Avg Temp': [25, 28, 32, 35, 43, 40,
                                35, 33, 30, 25, 20, 18]},
                  index=['Jan', 'Feb', 'Mar', 'Apr', 'May',
                         'Jun', 'Jul', 'Aug', 'Sep', 'Oct',
                         'Nov', 'Dec'])


# create stacked bar chart for monthly temperatures
df.plot(kind='bar', stacked=True, color=['red', 'skyblue', 'green'])

# labels for x & y axis
plt.xlabel('Months')
plt.ylabel('Temp ranges in Degree Celsius')

# title of plot
plt.title('Monthly Temperatures in a year')
```

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = '/content/drive/MyDrive/2016elections.csv'
election_df = pd.read_csv(file_path)

print("Columns in DataFrame:", election_df.columns)
print("Unique state values:", election_df['state'].unique())

states = ['Arkansas', 'Michigan', 'California', 'Wisconsin']
selected_states = election_df[election_df['state'].isin(states)].copy()

print("Filtered DataFrame:")
print(selected_states.head())


if not selected_states.empty:
    selected_states.set_index('state', inplace=True)
    vote_columns = ['pct_clinton', 'pct_trump', 'pct_johnson', 'pct_other']

    selected_states[vote_columns].plot(kind='bar', stacked=True,
                                       color=['blue', 'red', 'green', 'gray'],
                                       figsize=(8, 6))

    plt.xlabel("State")
    plt.ylabel("Vote Percentage")
    plt.title("2016 Election Vote Distribution for Arkansas, Michigan, California, and Wisconsin")
    plt.xticks(rotation=0)
    plt.legend(title="Candidate", labels=["Clinton", "Trump", "Johnson", "Others"])
    plt.show()
else:
    print("No data available for the specified states. Please verify the CSV content and column names.")
```
